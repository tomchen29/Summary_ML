```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams
sns.set(color_codes=True)
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
rcParams['figure.figsize'] = 8, 6
```

# Scatter Matrix (Pandas)

```python
pd.plotting.scatter_matrix(iris, figsize=(12,10));
```

# HeatMap (Seaborn)

```python
sns.heatmap(iris.corr(), annot=True);
```

# Pairplot (Seaborn)

```python
sns.pairplot(iris, hue='species', size=3, aspect=1);  # Aspect * size gives the width (in inches) of each facet.
# sns.pairplot(iris, hue="species",diag_kind="kde");
```

# Histogram (Pandas)

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12,8));
```

# Histogram (Seaborn)

```python
from scipy import stats
sns.distplot(iris.sepal_length, kde=False, fit=stats.norm)
```

# Box Plot (Pandas)

```python
iris.boxplot(by='species', figsize=(12,8));
```

# Bar Chart (Seaborn)

```python
sns.barplot(x="sepal_width", y="sepal_length", hue='species', data = iris)
```

# Pie Chart (Matplotlib)

```python
# Prepare the colors and labels
propotion = iris.species.value_counts() / iris.shape[0]
colors  = ['navy','lightcoral', 'turquoise']
labels = iris.species.unique()

plt.figure(figsize=(8,8))

## Plot the Pie Chart
patches, texts, autotexts = plt.pie(propotion,labels=labels,autopct='%1.1f%%',\
                            explode=[0.02 for _ in range(len(labels))], startangle=0, colors=colors)

plt.title('Sample Propotion of Iris Species',fontsize=30)

# Format the text labels (resize, change style)
for text in texts+autotexts:
    text.set_fontsize(20)
    text.set_fontweight('bold')

for text in autotexts:
    text.set_color('white')
    text.set_fontstyle('italic')
    
plt.show()
```

# Regression Plot (Seaborn)

```python
ax = sns.regplot(x="sepal_length", y="sepal_width", data=iris, fit_reg=True)
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
plt.show()
```

# Violin Plot (Seaborn)

```python
sns.violinplot(x='species', y='sepal_length', data=iris)
```
